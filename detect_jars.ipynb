```python
def timestamp_coords(img_width=1080):
   length = 900
   xmin = int((img_width - length)/2)
   ymin = 50
   xmax = xmin + length
   ymax = ymin + 50
   return [(xmin,ymin),(xmax,ymax)]

def draw_timestamp(draw,numitems):
   big_font = ImageFont.truetype("DejaVuSans.ttf",40)
   draw.rectangle(timestamp_coords(), fill="black", width=3)
   draw.text(timestamp_coords()[0]," camera 09      11/13/2018     item count: " + str(numitems), fill="white", font=big_font)
```

```python

num_items = 0
names=df_["Name"].unique()
names=sorted(names)
value=len(df_[df_.Name==names[0]])
decay=0.05
for name in names:

    with Image.open(frames + name) as im:
#         plt.figure()
#         plt.imshow(im)
        w, h = im.size
        df_im=df_[df_.Name==name]
        num = len(df_im)
        value=value*(1-decay)+num*decay

        print (num,round(value))
        for index,cur in df_im.iterrows():
            
            y1 = int(h *cur["bb_coord1"])  #y1
            x1 = int(w * cur["bb_coord2"]) #x1
            y2 = int(h * cur["bb_coord3"]) #y2
            x2 = int(w * cur["bb_coord4"]) #x2
            font = ImageFont.truetype("DejaVuSans.ttf", 30)
            draw = ImageDraw.Draw(im)
            draw_timestamp(draw,int(round(value)))
            draw.rectangle([(x1,y1-30), (x1+250,y1)], fill=255, width=15)
            draw.rectangle([(x1,y1), (x2,y2)], outline=255, width=15)
            draw.text((x1, y1-30), "Jar", fill=0, font=font)
        del draw
        
#         plt.figure()
#         plt.imshow(im)
        im.save(res + name)
```
